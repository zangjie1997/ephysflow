```python
def filter_syllables(all_syll, syl_type=None, test_frac=0):
    if syl_type:
        syl_selection = (all_syll['syl_type']==syl_type) & (all_syll['keep']==True)
    else:
        syl_selection = (all_syll['keep']==True)
        
    select_syl = all_syll[syl_selection]
    total_trials = select_syl.shape[0]
    test_trials = np.int(test_frac*total_trials)
    
    syl_train = select_syl[:total_trials-test_trials].reset_index(drop=True)
    syl_test = select_syl[-test_trials:]
    return syl_train, syl_test

def keras_data_set(syl_df, fit_pars, units_list, flat_features=False):
    # all syl and units in unts_list are from the same bird and sess
    # (they share the kwd and kwik file)
    s_f = units_list[0].sampling_rate
    kw_file = units_list[0].h5_file
    ms_to_samp = s_f//1000.
    
    syl_starts = (syl_df.start.as_matrix() * ms_to_samp 
                  + syl_df.mot_start.as_matrix()).astype(np.int)
    syl_recs = syl_df.rec.as_matrix().astype(np.int)
    # the starts in absolute rel. to kw(k/d) file
    syl_starts_abs = kwkf.apply_rec_offset(kw_file, syl_starts, syl_recs)
    
    s_v = []
    target = []
    logger.info('Collecting neural features and target')
    for i_syl, syl in tqdm(syl_df.iterrows(), total=syl_df.shape[0]):
        #print(i_syl)
        len_ms = int(syl['duration'])
        len_samples = int(syl['duration'] * ms_to_samp)
        pars_ms = np.stack([np.array(syl[p_name]) for p_name in ['alpha', 'beta', 'env']])[:,:len_ms]
        pars_ms = fit_pars['par_reg_fcn'](pars_ms)
        model_pars = bp.col_binned(pars_ms, fit_pars['bin_size'])/fit_pars['bin_size']
        this_sv, ths_sv_u = unitobjs.support_vector(np.array([syl_starts_abs[i_syl]]), 
                                             len_samples, 
                                             units_list,
                                            bin_size = fit_pars['bin_size'],
                                            history_bins = fit_pars['history_bins']+1,
                                            no_silent=False)
        s_v.append(this_sv)
        target.append(model_pars)

    logger.info('stacking')
    fv_list = []
    tv_list = []
    for fv, tv in tqdm(zip(s_v, target)):
        f_win, t_win = nd.one_frame_to_window(fv, tv, fit_pars['history_bins'])
        fv_list.append(f_win)
        tv_list.append(t_win)
    feature_window = np.vstack(fv_list)
    target_window = np.vstack(tv_list)
    
    logger.info('Making Keras datasets')
    total_samples = feature_window.shape[0]
    test_samples = int(fit_pars['test_fraction']*total_samples)

    x = feature_window[:-test_samples, :, :]
    Y = target_window[:-test_samples, :]

    x_t = feature_window[-test_samples: , :, :]
    Y_t = target_window[-test_samples: , :]
    logger.info('Done, train shapes are {0}, {1}'.format(x.shape, Y.shape))
    
    if flat_features:
        logger.info('Flattening the feature dimensions (for rrnn in keras, as opposed to lstm)')
        X = x.reshape([x.shape[0], -1])
        X_t = x_t.reshape([x_t.shape[0], -1])
    else:
        X = x
        X_t = x_t
    
    return X, Y, X_t, Y_t


class sessData():
    def __init__(self, bird, sess, fit_pars):
        self.bird = bird
        self.sess = sess
        self.fp = fit_pars
        self.all_syl = None
        
        self.X = None
        self.X_t = None
        self.Y = None
        self.Y_t = None
        
        # load metadatas and files
        self.fn = et.file_names(self.bird, self.sess)
        self.exp_par = et.get_parameters(bird, sess)

        self.kwik_file = et.open_kwik(bird, sess)
        self.kwd_file = et.open_kwd(bird, sess)
        
        self.load_syl()
        self.load_units()
        
    def load_syl(self):
        logger.debug('Loading syllable data for {0}-{1} with file descr {2}'.format(bird, sess, self.fp['syl_file']))
        loaded_syl = pd.concat([pd.read_pickle(os.path.join(self.fn['folders']['ss'], 
                                                     '{0}_{1}.pickle'.format(self.fp['syl_file'],
                                                                             syl_type)))
                         for syl_type in self.fp['syl_types']]).sort_values(by='mot_id')
        
        self.all_syl, _ = filter_syllables(loaded_syl)
        
    def load_units(self):
        logger.debug('Loading units data for {0}-{1} with type {2}'.format(bird, sess, self.fp['unit_type']))
        u_type = self.fp['unit_type']
        if u_type is 'cluster':
            all_sess_units = um.list_sess_units(self.bird, self.sess)
            all_units = kwkf.list_units(kwik_file, group=0, sorted=False)
            self.units_list = [unitobjs.Unit(clu, kwik_file=self.kwik_file) for clu in all_units.clu]

        elif u_type is 'threshold':     
            chans_list = np.array(self.exp_par['channel_config']['neural'])
            self.units_list = [unitobjs.threshUnit(ch, kwd_file=self.kwd_file) for ch in chans_list]
            
        else:
            raise NotImplementedError('dont know how to handle unit type {}'.format(u_type))
            
        logger.debug('Loaded {} units'.format(len(self.units_list)))
        
    def gimme_set(self, test_trials=None):
        flat_features = True if self.fp['network'] is 'ffnn' else False
        self.X, self.Y, self.X_t, self.Y_t = keras_data_set(self.all_syl, self.fp, 
                                                            self.units_list, 
                                                            flat_features=flat_features)
        return self.X, self.Y, self.X_t, self.Y_t
```

## Train the network

```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout, Input
from keras.layers import LSTM, Activation, advanced_activations
from keras.regularizers import l2
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.models import Model
```

```python
fit_pars = {'bin_size': 3,
           'history_bins': 20,
           'latency': 0,
           'test_fraction': 0.1,
            'num_ep': 30,
           'par_reg_fcn': syn.regularize_pars,
           'network': 'ffnn',
           'syl_file': 'syllable_pandas_template',
           'syl_types': ['0', '-1'],
           'unit_type': 'cluster'}

sess_data = sessData(bird, sess, fit_pars)
X, Y, X_t, Y_t = sess_data.gimme_set()
```

```python
X.shape
```

```python
# attempt to make a 3 column network for all parameters
n_feat = X.shape[-1]
n_test = 10
l2_val = 0.001
dropout_val = 0.2
cross_valid = False
num_neurons = [20, 30]
valid_split = 0.1
n_feat = X.shape[-1]

neural_input = Input(shape=(n_feat, ), name='neural_input')

x = Dense(n_feat//2, kernel_initializer='uniform', activation='relu')(neural_input)
x = Dense(n_feat//4, kernel_initializer='uniform', activation='relu')(x)
x = Dense(n_feat//16, kernel_initializer='uniform', activation='relu')(x)

y = Dense(n_feat//2, kernel_initializer='uniform', activation='relu')(neural_input)
y = Dense(n_feat//4, kernel_initializer='uniform', activation='relu')(y)
y = Dense(n_feat//16, kernel_initializer='uniform', activation='relu')(y)

z = Dense(n_feat//2, kernel_initializer='uniform', activation='relu')(neural_input)
z = Dense(n_feat//4, kernel_initializer='uniform', activation='relu')(z)
z = Dense(n_feat//16, kernel_initializer='uniform', activation='relu')(z)

alpha_output = Dense(1, kernel_initializer='normal', name='alpha_output')(x)
beta_output = Dense(1, kernel_initializer='normal', name='beta_output')(y)
env_output = Dense(1, kernel_initializer='normal', name='env_output')(z)


model = Model(inputs=[neural_input], 
             outputs=[alpha_output, beta_output, env_output])

model.compile(optimizer='rmsprop',
              loss={'alpha_output': 'mean_absolute_error',
                   'beta_output': 'mean_absolute_error', 
                   'env_output': 'mean_absolute_error'},
              loss_weights={'alpha_output': .5,
                   'beta_output': 1., 
                   'env_output': 1.})


model_name = 'ffnn_custom_warptemp_bs{0:02d}_hs{1:02d}_te{2:03d}.h5'.format(fit_pars['bin_size'], 
                                                                     fit_pars['history_bins'], 
                                                                     fit_pars['num_ep'])
model_path = os.path.join(fn['folders']['ss'], 'model_' + model_name)
weights_path = os.path.join(fn['folders']['ss'], 'weights_' + model_name)
callbacks = [EarlyStopping(monitor='val_loss', patience=4, verbose=1),
             ModelCheckpoint(filepath=model_path, monitor='val_loss', 
                             save_best_only=True, 
                             verbose=1)
            ]
model.fit({'neural_input': X},
         {'alpha_output': Y[:,0], 'beta_output': Y[:, 1], 'env_output': Y[:, 2]},
         epochs=fit_pars['num_ep'],
         batch_size=10,
        validation_split=valid_split,
         callbacks=callbacks)

```
